```python
cap=cv2.VideoCapture(3)
while cap.isOpened():
  ret,frame=cap.read()
  cv2.imshow("Movenet Multipose",frame)
  if cv2.waitkey(10) & 0xFF==ord('q'):
    break

cap.release()
cv2.destroyAllWindows()
```
